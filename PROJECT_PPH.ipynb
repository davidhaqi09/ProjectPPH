## Data cleaning

### Subtask:
Clean the data in `df_raw`.


**Reasoning**:
Clean the data in `df_raw` according to the instructions, including handling missing values, inconsistent data types, duplicates, outliers, and inconsistent values. Then, save the cleaned DataFrame to `df_cleaned`.



```python
# Handle missing values
missing_percentage = df_raw.isnull().sum() * 100 / len(df_raw)
columns_to_drop = missing_percentage[missing_percentage > 60].index
df_cleaned = df_raw.drop(columns=columns_to_drop)

# Impute missing values in remaining columns
for col in df_cleaned.select_dtypes(include=['number']).columns:
    if df_cleaned[col].isnull().any():
      df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].median()) # Using median for numerical features

# Address inconsistent data types
df_cleaned['tahun'] = pd.to_numeric(df_cleaned['tahun'], errors='coerce')
df_cleaned = df_cleaned.dropna(subset=['tahun'])
df_cleaned['tahun'] = df_cleaned['tahun'].astype(int)


# Remove duplicates
df_cleaned = df_cleaned.drop_duplicates()

# Handle outliers (using IQR method as an example)
for col in df_cleaned.select_dtypes(include=['number']).columns:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned[col] = df_cleaned[col].clip(lower=lower_bound, upper=upper_bound)

# Handle inconsistent values (example for 'tax_holiday_status')
df_cleaned['tax_holiday_status'] = df_cleaned['tax_holiday_status'].astype(str).str.strip().str.lower() # Convert to string, strip whitespace, lowercase


display(df_cleaned.head())
display(df_cleaned.info())
```

## Data wrangling

## Model training

### Subtask:
Train a time series forecasting model to predict future financial performance.


**Reasoning**:
I will use Prophet model to forecast the financial performance. Prophet is robust to seasonality and trend changes. I will use 'pendapatan_operasional' as the target variable. I will prepare the data by selecting the 'tahun' and 'pendapatan_operasional' columns and renaming them to 'ds' and 'y', which are required by Prophet. Then, I will split the data into training and testing sets. Finally, I will train the Prophet model on the training data, make predictions on the test data and future periods, and store them in a dataframe.



```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Prepare the data for Prophet
df_prophet = df_cleaned[['tahun', 'pendapatan_operasional']].rename(columns={'tahun': 'ds', 'pendapatan_operasional': 'y'})

# Split the data into training and testing sets
train_data = df_prophet[:-2]  # Use all but the last two years for training
test_data = df_prophet[-2:]   # Use the last two years for testing


# Initialize and fit the Prophet model
model = Prophet()
model.fit(train_data)

# Make predictions for the test data
future = model.make_future_dataframe(periods=2, freq='Y') # predict next 2 years
forecast = model.predict(future)

# Evaluate the model
y_true = test_data['y'].values
y_pred = forecast['yhat'][-2:].values
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Make predictions for future periods (2028 and beyond)
future_periods = 3 # predict next 3 years
future_forecast = model.make_future_dataframe(periods=future_periods, freq='Y')
future_forecast = model.predict(future_forecast)

# Store predictions in a DataFrame
predictions = future_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
predictions = predictions.rename(columns={'ds': 'year', 'yhat': 'predicted_revenue'})
display(predictions.tail())
```

```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd

# Prepare the data for Prophet
df_prophet = df_cleaned[['tahun', 'pendapatan_operasional']].rename(columns={'tahun': 'ds', 'pendapatan_operasional': 'y'})
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'], format='%Y')

# Split the data into training and testing sets
train_data = df_prophet[:-2]  # Use all but the last two years for training
test_data = df_prophet[-2:]   # Use the last two years for testing

# Initialize and fit the Prophet model
model = Prophet()
model.fit(train_data)

# Make predictions for the test data
future = model.make_future_dataframe(periods=2, freq='Y')
forecast = model.predict(future)

# Evaluate the model
y_true = test_data['y'].values
y_pred = forecast['yhat'][-2:].values
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Make predictions for future periods (2028 and beyond)
future_periods = 3
future_forecast = model.make_future_dataframe(periods=future_periods, freq='Y')
future_forecast = model.predict(future_forecast)

# Store predictions in a DataFrame
predictions = future_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
predictions = predictions.rename(columns={'ds': 'year', 'yhat': 'predicted_revenue'})
display(predictions.tail())
```
